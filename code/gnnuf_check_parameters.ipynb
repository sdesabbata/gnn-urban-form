```python
# Torch
from torch_geometric.nn import GAE
# GNN models and utils
from gnnuf_models import *
# OS environment setup
from local_directories import *
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
if torch.cuda.is_available():
    print("Run on CUDA")
else:
    print("Run on CPU")
```

```python
# Load model
model_name = "gnnuf_model_v0-5"
model = GAE(GINEEncoder(1, 1, 64, 2))
model.load_state_dict(torch.load(this_repo_directory + "/models/" + model_name + ".pt", map_location=device))
model = model.to(device)
model.eval()
```

```python
total_num_of_param = 0
for name, param in model.named_parameters():
    if param.requires_grad:
        print(f"{name=}, len={len(param.data)}\n{param.data}\n\n")
        total_num_of_param += len(param.data)
```
